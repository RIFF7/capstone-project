### Analisa Kejadian Kejahatan Hari dalam Seminggu

```python
# Ekstrak hari dalam seminggu dari kolom tanggal kejadian kejahatan (0 = Senin, 1 = Selasa, ..., 6 = Minggu)
data_clean['Hari_Occ'] = data_clean['TglOcc'].dt.dayofweek

# Mengganti angka dengan nama hari
nama_hari = {0: 'Senin', 1: 'Selasa', 2: 'Rabu', 3: 'Kamis', 4: 'Jumat', 5: 'Sabtu', 6: 'Minggu'}
data_clean['Hari_Occ'] = data_clean['Hari_Occ'].map(nama_hari)

# Menghitung jumlah kejahatan untuk setiap hari dalam seminggu
crime_by_day = data_clean.groupby('Hari_Occ').size()

# Plot distribusi jumlah kejahatan per hari dalam seminggu
plt.figure(figsize=(10, 6))
crime_by_day.plot(kind='bar', color='skyblue')
plt.title('Distribusi Kejahatan per Hari dalam Seminggu')
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Jumlah Kejahatan')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
