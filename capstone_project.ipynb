```python
# Mengetahui isi data dari jumlah angka minus diatas

# Mengelompokkan data berdasarkan 'DescKej'
grouped_data = data_crime[data_crime['UsiaKor'] < 0].groupby('DescKej')

# Membuat list kosong untuk menyimpan hasil
rows = []

# Mengisi list dengan umur yang memiliki nilai negatif
for descKej, group in grouped_data:
    umur_negatif = group[group['UsiaKor'] < 0]['UsiaKor'].unique()
    rows.append({'Deskripsi Kejahatan': descKej, 'Umur Negatif': umur_negatif})

# Membuat DataFrame dari list hasil
umur_minus_descKej = pd.DataFrame(rows)

# Menghilangkan indeks saat menampilkan DataFrame
print(umur_minus_descKej.to_string(index=False))
```
