```python
# 1. Tren Waktu Kejadian
data_crime['TglOcc'] = pd.to_datetime(data_crime['TglOcc'])
data_crime.set_index('TglOcc', inplace=True)
daily_crime_count = data_crime.resample('D').size()
plt.plot(daily_crime_count)
plt.title('Tren Harian Kejahatan')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kejahatan')
plt.show()
```

```python
# 2. Distribusi Kejadian Berdasarkan Area
area_crime_count = data_crime['NmArea'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=area_crime_count.index, y=area_crime_count.values)
plt.title('Distribusi Kejahatan Berdasarkan Area')
plt.xlabel('Area')
plt.ylabel('Jumlah Kejahatan')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Jenis Kejahatan yang Umum
jenis_kejahatan_count = data_crime['DescKej'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=jenis_kejahatan_count.index, y=jenis_kejahatan_count.values)
plt.title('Jenis Kejahatan yang Umum')
plt.xlabel('Jenis Kejahatan')
plt.ylabel('Jumlah Kejahatan')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Profil Korban
plt.figure(figsize=(10, 6))
sns.histplot(data_crime['UsiaKor'], bins=20, kde=True)
plt.title('Distribusi Usia Korban Kejahatan')
plt.xlabel('Usia')
plt.ylabel('Jumlah Kejahatan')
plt.show()
```

```python
# 5. Status Laporan dan Deskripsi Status
status_laporan_count = data_crime['DescStatLpr'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=status_laporan_count.index, y=status_laporan_count.values)
plt.title('Distribusi Status Laporan Kejahatan')
plt.xlabel('Status Laporan')
plt.ylabel('Jumlah Kejahatan')
plt.xticks(rotation=45)
plt.show()
```

```python
# Korelasi Antar Kolom
correlation= data_crime.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='RdYlBu', fmt='.2f')
plt.title('Korelasi Setiap Kolom')
plt.show()
```
