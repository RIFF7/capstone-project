## Proses Mengambil Value Kolom UsiaKor & JenisKelKor

- Mengambil Data pada kolom UsiaKor sama dengan 0 dan juga kurang dari 0 (Nilai Minus)
- Mengambil Data pada kolom JenisKelKor dengan values [X, H, NaN]

### UsiaKor

```python
# Menghitung kemunculan data pada kolom UsiaKor
usia_kor_counts = data_crime['UsiaKor'].value_counts(dropna = False)
usia_kor_counts
```

```python
# Melakukan pengecekan pada kolom UsiaKor apakah ada data kosong
data_usia_null = data_crime[data_crime['UsiaKor'].isnull()]

# Menampilkan hasil
if data_usia_null.empty:
    print("Tidak ada data yang kosong pada kolom UsiaKor")
else:
    print("Data kosong pada kolom UsiaKor:")
    print(data_usia_null)
```

```python
# Pengecekan kolom UsiaKor
usia_kor_counts = data_crime['UsiaKor'].value_counts()

# Menampilkan jumlah kemunculan angka 0 dan angka negatif
count_zero = usia_kor_counts.get(0, 0)
count_minus = usia_kor_counts.loc[usia_kor_counts.index < 0].sum()

# Menampilkan hasil rule
print("Jumlah Angka 0 pada kolom UsiaKor:", count_zero)
print("Jumlah Angka Negatif pada kolom UsiaKor:", count_minus)
```

### JenisKelKor

```python
# Menampilkan jumlah nilai Unik pada kolom JenisKelKor
jumlah_jenis_kelKor = data_crime['JenisKelKor'].nunique()

# Tampilkan hasil
print("Jumlah Nilai Unik pada kolom JenisKelKor:", jumlah_jenis_kelKor)
```

```python
# Mengambil Nilai Unik pada kolom JenisKelKor
daftar_unik_kelKor = data_crime['JenisKelKor'].unique()

# Tampilkan hasil dalam bentuk array
daftar_unik_kelKor
```

```python
# Memfilter data berdasarkan kondisi yang diberikan
filtered_data = data_crime[(data_crime['UsiaKor'] == 0) | (data_crime['UsiaKor'] < 0) & (data_crime['JenisKelKor'].isin(['X', 'H', '-']))]

# Menampilkan hasil
unknown_data = filtered_data
unknown_data
```
